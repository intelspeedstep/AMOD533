```
# Convert to pandas DataFrame
import io
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/intelspeedstep/AMOD533/master/UCI_Credit_Card.csv')
print(df.info(verbose=True))
df.head()
```

### -------------------------------------------Question 1 Data Transformation: -----------------------------------------
-----------------------------------------------------

```
# I pick AGE variable as it is a numeric and in int64 format, and convert it into a categorical(object) variable in python
print(df.describe()['AGE'])
df['AGE_CAT']=df['AGE'].astype(object)
df['AGE_CAT'].dtype
```

### --------------------------------------Question 2 Data Substitution:---------------------------------------------

```
# Check the value count of the MARRIAGE attributes as well as its distributions
print(df.shape)
print(df.MARRIAGE.value_counts(dropna=False))
df.MARRIAGE.value_counts(dropna=False, normalize=True)
```

```
# Convert the numeric variable to categorical variable. Since the MARRIAGE attributes has 3 levels, I replaced it with the map function to convert the numeric values to object, which is categorical in python. 
df['MARRIAGE']=df.MARRIAGE.map({1:'married',2:'single',3:'others'})
print(df['MARRIAGE'].value_counts(dropna=False))
print(df['MARRIAGE'].value_counts(dropna=False, normalize=True))
df['MARRIAGE'].dtype
```

### ------------------------------Question 3 Data Reduction:----------------------------------------

```
# Binning using pd.qcut() in pandas
# First, I convert it into a categorical variable with 10 values that give data 10% for with each category
# Then, assign both the values and description in to 'AGE_CAT' and 'AGE_CAT_DESC' to ranges of conversion
# So, I get categorical variable for age as well the description of the each category 
print(df.describe()['AGE'])
df['AGE_CAT']=pd.qcut(df['AGE'],10, labels=False, precision=0)
df['AGE_CAT_DESC']=pd.qcut(df['AGE'],10, precision=0)
df[['AGE_CAT','AGE_CAT_DESC']].drop_duplicates().sort_values(by=['AGE_CAT']).reset_index(drop=True)
```

```
# Bucketing using setting category threholds on my own
# First, I need to check the data distribution for age
# Then, assign both the values and description variable on my own
print(df.describe(percentiles=[.1,.25,.5,.75,.9])['AGE'])
df.loc[df['AGE']<=25,'AGE_BUCKETING']='21-25'
df.loc[(df['AGE']>25) & (df['AGE']<=28),'AGE_BUCKETING']='26-28'
df.loc[(df['AGE']>28) & (df['AGE']<=34),'AGE_BUCKETING']='29-34'
df.loc[(df['AGE']>34) & (df['AGE']<=41),'AGE_BUCKETING']='35-41'
df.loc[(df['AGE']>41) & (df['AGE']<=49),'AGE_BUCKETING']='42-49'
df.loc[df['AGE']>49,'AGE_BUCKETING']='50-79'
df['AGE_BUCKETING_CODE']=df['AGE_BUCKETING'].map({'21-25':0,'26-28':1,'29-34':2,'35-41':3,'42-49':4,'50-79':5})
df[['AGE_BUCKETING_CODE','AGE_BUCKETING']].drop_duplicates().sort_values(by=['AGE_BUCKETING_CODE']).reset_index(drop=True)
```
