### -------------------------------------------Question 1 Determine what correlations there are among independent variables (= predictors) in the dataset by setting up a correlation table.  Which is the most important predictor variable and why? -----------------------------------------

```
# I would pick limit_bal variable as the target variable for my linear regression since limit_bal is a continuous variable
# first, I would generate a correclation matrix among all the indepentdent variables as well as the target variables
# then, exclude the ID variable. No need to do correlation on ID
# Base on the results below, we can see that 1) both PAY and BILL_AMT5 variables is most important variables in my linear regression model
# Those are important variables since their correlation to the target variables is 0.295562 are much higher than that to the other variables.
df.drop(['ID'], axis=1).corr().style.background_gradient(cmap='cool')
```

### --------------------------------------Question 2 Build a simple linear regression based on this predictor and evaluate the quality of your model.---------------------------------------------

```
# I set Limit_bal as the target variable and others as predictors.
# fit the model using sklean.LinearRegression()
# Output the beta coefficient and intercept, as well as the R square on on the training datasets.
# As you can see R-square is 0.09, the prediction power of using single variable in this case is very weak.
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression().fit(df.BILL_AMT5.values.reshape(-1,1),df.LIMIT_BAL)
print("linear_model.coef_:", linear_model.coef_)
print("linear_modelintercept_:", linear_model.intercept_)
print("R-square score: {:.2f}".format(linear_model.score(df.BILL_AMT5.values.reshape(-1,1), df.LIMIT_BAL)))
```

### ------------------------------Question 3 Set up up a best subset regression table and check how many of the variables should be combined in a multiple linear regression, then build and run your multiple regression model.----------------------------------------


```
# First, we need to avoid any multi-collinearity, meaning we need to keep one variable if many of them are highly correlated.  
# So, in this case, I only keep PAY_2 from the PAY variables, BILL_AMT5 from BILL variables, and other predictors in the dataset
# Then, we fit in a multiple linear regression. The R-square is 0.32, meaning the prediction of the subset variables are much better than a single predictor.
from sklearn.linear_model import LinearRegression
X=df.drop(['PAY_0','PAY_3','PAY_4','PAY_5','PAY_6','BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT6','LIMIT_BAL'],axis=1)
linear_model = LinearRegression().fit(X,df.LIMIT_BAL)
print("linear_model.coef_:", linear_model.coef_)
print("linear_modelintercept_:", linear_model.intercept_)
print("R-square score: {:.2f}".format(linear_model.score(X, df.LIMIT_BAL)))
```

### ------------------------------Question 4 Build and run a logistic regression model. What must the data type of your target variable be to make this work.----------------------------------------

```
# For the logistic regression, the target variable must be a binary variable, meaning must be 1 or 0.
# The target variable in this case is 'default.payment.next.month'. 
# To avoid any multi-collinearity, we only keep PAY_0 from the PAY variables, and BILL_AMT0 from BILL variables, and keep other predictors in the dataset
# 
from sklearn.linear_model import LogisticRegression
X=df.drop(['PAY_2','PAY_3','PAY_4','PAY_5','PAY_6','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6','default.payment.next.month'],axis=1)
logit_model = LogisticRegression(max_iter=1000).fit(X,df['default.payment.next.month'])
print("linear_model.coef_:", logit_model.coef_)
print("linear_modelintercept_:", logit_model.intercept_)
print("Model accuracy is : {:.2f}".format(logit_model.score(X, df['default.payment.next.month'])))
```
